```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib
```

```python
cislo = 21                      # volim podle seznamu

e4 = 4 * 10**(-3)               # [m]
E = 200000 * 10**6              # [Pa]
A = 70 * 10**(-6)               # [m^2]
F = 4000+150*cislo              # [N]
beta = np.radians(30)           # []
a = (600 + 20*cislo) * 10**(-3)   # [m]

''''''''''''''''''''''''''''''''''''''

print("Velikost F [N]: ",F)                    # sila F
print("Velikost a [mm]: ", a*10**(3))            # delka L3


'''VOLIM DRUHOU KONSTRUKCI'''

'''DOPOCITANI DELEK L1, L2, L4'''

L1 = a
L2 = a*np.sin(np.pi/2 - beta)/np.sin(beta)
L3 = a
L4 = np.sqrt((2*a)**2 + L2**2)

print()
print("Velikost L1 [m]: ", L1)
print("Velikost L2 [m]: ", L2)
print("Velikost L3 [m]: ", L3)
print("Velikost L4 [m]: ", L4)   



''''''''''''''''''''''''''''''                  # vypocet reakci ve vazbach a normalovych sil v prutech

alpha = np.arcsin(2*a/L4)
b = L2

A1 = np.array([[0, 1, 0, -np.cos(alpha)],       # matice A1
              [1, 0, 1, np.sin(alpha)], 
              [0, 0, b, b*np.sin(alpha) - a*np.cos(alpha)],
              [L1*a*np.cos(alpha)/(E*A*b), -L2*np.cos(alpha)/(E*A), L3/(E*A)*(np.sin(alpha)-a*np.cos(alpha)/b), -L4/(E*A)]])


g = np.array([[0],                              # matice g
              [0],
              [-2*F*a*np.cos(beta)],
              [e4]])


Ainv = np.linalg.inv(A1)                        # inverzni matice k matici A1



Z = np.dot(Ainv,g)                              # matice reseni Z, vznikla vynasobenim matic A^(-1) * b

print()
print("N_1 [N]: ", float(Z[0]))                        # vypise vysledky
print("N_2 [N]: ", float(Z[1]))
print("N_3 [N]: ", float(Z[2]))
print("X [N]: ", float(Z[3]))



''''''''''''''''''''''''''''''                  # vypocet posuvu pomoci Castiglianovy vety

'''UZEL P:'''

'''POSUV VE SMERU VODOROVNEM u_P:'''
    
N1 = float(Z[0])
N2 = float(Z[1])
N3 = float(Z[2])


uP = 1/(E*A)*(N1*L1*a/b - N2*L2 - N3*L3*a/b)

print()
print("Posuvy uzlu P:")
print("u_P [mm]: ", uP*10**3)                        # posuv bodu P v horizontalnim smeru


'''POSUV VE SMERU VERTIKALNIM v_P:'''

vP = N3*L3/(A*E)

print("v_P [mm]: ", vP*10**3)                        # posuv bodu P ve vertikalnim smeru
```
