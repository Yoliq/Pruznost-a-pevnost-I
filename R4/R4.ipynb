```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib
```

```python
cislo = 21 # volim podle seznamu

E = 200000 * 10**6              # [Pa]
alpha = 1.2 * 10**(-5)          # [K^(-1)]
deltaT4 = 100+cislo             # [°C]
A = 80 * 10**(-6)               # [m^2]
F = 4000+150*cislo              # [N]
beta = np.radians(30)           # []
L3 = (800+20*cislo) * 10**(-3)  # [m]

''''''''''''''''''''''''''''''''''''''

print("Velikost F [N]: ",F)                    # sila F
print("Velikost L3 [m]: ", L3)                 # delka L3
print("Velikost deltaT4 [K^(-1)]: ", deltaT4)  # zmena teploty prutu cislo 4


'''VOLIM DRUHOU KONSTRUKCI'''

'''DOPOCITANI DELEK L1, L2, L4'''
gama = np.pi/2 - beta

L2 = L3/np.tan(gama)

L1 = np.sqrt(L3**2 + L2**2)

L4 = L1

print()
print("Velikost L2 [m]: ", L2)
print("Velikost L1 [m]: ", L1)
print("Velikost L4 [m]: ", L4)   



''''''''''''''''''''''''''''''                  # vypocet reakci ve vazbach a normalovych sil v prutech

A1 = np.array([[ np.sin(gama), 0, 0, 0, 1, 0, 0, 0, 0],  # matice A
              [np.cos(gama), 1, 0, 0, 0, 1, 0, 0, 0], 
              [0, 0, 1, np.cos(beta), 0, 0, 0, 0, 0],
              [0, 1, 0, np.sin(beta), 0, 0, 0, 0, 0],
              [np.cos(beta), 0, 1, 0, 0, 0, np.cos(gama), 0, 0],
              [np.sin(beta), 0, 0, 0, 0, 0, -np.sin(gama), 0, 0],
              [0, 0, 0, np.cos(beta), 0, 0, 0, 0, 1],
              [0, 0, 0, np.sin(beta), 0, 0, 0, 1, 0],
              [-L1*np.cos(beta)/(E*A*(np.cos(beta)+((np.sin(beta))**2)/np.cos(beta))), L2*np.sin(beta)/(E*A), L3*np.cos(beta)/(E*A), -L4/(E*A), 0, 0, 0, 0, 0]])


b = np.array([[0],                              # matice b
              [0],
              [3*F*np.sin(beta)],
              [-3*F*np.cos(beta)],
              [0],
              [0],
              [0],
              [0],
              [L4*alpha*deltaT4]])


Ainv = np.linalg.inv(A1)                        # inverzni matice k matici A



X = np.dot(Ainv,b)                              # matice reseni X, vznikla vynasobenim matic A^(-1) * b


print()
print("N_1 [N]: ", float(X[0]))                        # vypise vysledky
print("N_2 [N]: ", float(X[1]))
print("N_3 [N]: ", float(X[2]))
print("N_4 [N]: ", float(X[3]))
print("X_H [N]: ", float(X[4]))
print("Y_H [N]: ", float(X[5]))
print("P_p [N]: ", float(X[6]))
print("X_Z [N]: ", float(X[7]))
print("Y_Z [N]: ", float(X[8]))



''''''''''''''''''''''''''''''                  # vypocet posuvu pomoci Castiglianovy vety

'''UZEL B:'''

'''POSUV VE SMERU PUSOBENI SILY 1:'''
    
N_4 = float(X[3])

u_B = L4*(N_4*1/(E*A) + alpha*deltaT4)

print()
print("u_B [mm]: ", u_B*10**3)                        # posuv bodu B ve smeru sily 1, u_B
```
